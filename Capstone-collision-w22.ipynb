```python
# Remove unnecessary data [Drop these columns]
df1 = df.drop(columns = ['OBJECTID', 'SEVERITYCODE.1', 'REPORTNO', 'INCKEY', 'COLDETKEY', 
              'X', 'Y', 'STATUS','ADDRTYPE',
              'INTKEY', 'LOCATION', 'EXCEPTRSNCODE',
              'EXCEPTRSNDESC', 'SEVERITYDESC', 'INCDATE',
              'INCDTTM', 'JUNCTIONTYPE', 'SDOT_COLCODE',
              'SDOT_COLDESC', 'PEDROWNOTGRNT', 'SDOTCOLNUM',
              'ST_COLCODE', 'ST_COLDESC', 'SEGLANEKEY',
              'CROSSWALKKEY', 'HITPARKEDCAR', 'PEDCOUNT', 'PEDCYLCOUNT',
              'PERSONCOUNT', 'VEHCOUNT', 'COLLISIONTYPE',
              'SPEEDING', 'UNDERINFL', 'INATTENTIONIND'])
```

```python
# Convert column to category
df1["WEATHER"] = df1["WEATHER"].astype('category')
df1["ROADCOND"] = df1["ROADCOND"].astype('category')
df1["LIGHTCOND"] = df1["LIGHTCOND"].astype('category')

# Assign variable to new column for analysis
df1["WEATHER_CAT"] = df1["WEATHER"].cat.codes
df1["ROADCOND_CAT"] = df1["ROADCOND"].cat.codes
df1["LIGHTCOND_CAT"] = df1["LIGHTCOND"].cat.codes

df1.head(5)
```

```python
from sklearn.utils import resample
```

```python
# Seperate majority and minority classes
df1_one = df1[df1.SEVERITYCODE==1]
df1_two = df1[df1.SEVERITYCODE==2]

#Downsample majority class
df1_one_downsampled = resample(df1_one, replace=False, n_samples=58188, random_state=123)

# Combine minority class with downsampled majority class
df_balanced = pd.concat([df1_one_downsampled, df1_two])

# Display new class counts
df_balanced.SEVERITYCODE.value_counts()
```

Methodology : Our data is now ready to be fed into different machine learning models. 

```python
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
X = np.asarray(df_balanced[['WEATHER_CAT', 'ROADCOND_CAT', 'LIGHTCOND_CAT']])
X[0:5]
```

```python
Y = np.asarray(df_balanced['SEVERITYCODE'])
Y[0:5]
```

```python
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=4)
print ('Train set:', X_train.shape,  Y_train.shape)
print ('Test set:', X_test.shape,  Y_test.shape)
```

K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = 25
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,Y_train)
neigh

Kyhat = neigh.predict(X_test)
Kyhat[0:5]
```

```python
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import f1_score
```

```python
jaccard_similarity_score(Y_test, Kyhat)
```

```python
f1_score(Y_test, Kyhat, average='macro')
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
df_Tree = DecisionTreeClassifier(criterion="entropy", max_depth = 7)
df_Tree
df_Tree.fit(X_train,Y_train)
```

```python
DTyhat = df_Tree.predict(X_test)
print (DTyhat[0:5])
print (Y_test[0:5])
```

```python
jaccard_similarity_score(Y_test, DTyhat)
```

```python
f1_score(Y_test, DTyhat, average='macro')
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
import time
from sklearn.metrics import accuracy_score, log_loss, jaccard_similarity_score, classification_report
from sklearn.metrics import precision_score, recall_score, roc_curve
```

```python
model_rf = RandomForestClassifier(n_estimators=10, max_features=3, max_depth =12,criterion='entropy',random_state=0, n_jobs=-1)
model_rf.fit(X_train,Y_train)
yhat_rf = model_rf.predict(X_test)
jaccard_rf = jaccard_similarity_score(Y_test,yhat_rf)
c_rf = classification_report(Y_test,yhat_rf)
prec_rf = precision_score(Y_test, yhat_rf)
rec_rf = recall_score(Y_test, yhat_rf)
print('Jaccard :',jaccard_rf,'\n', c_rf)
```
